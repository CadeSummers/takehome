## This is a file for the analysis of ambulance diversion hrs from facilities with pediatric vs non-pediatric trauma centers

```python
#2
import pandas as pd
import numpy as np
from time import time
from matplotlib import pyplot as plt
from excel_analysis import b_ambulance_diversion_hours

data = b_ambulance_diversion_hours()

pediatric_trauma_df = data[0]
non_pediatric_trauma_df = data[1]

plt.title("EMS Ambulance Diversion Hours Compared Between\nPediatric Trauma and Non-Pediatric Trauma Center Facilities, YoY \n", fontdict={"fontsize" : 16})

plt.xlabel("Years since 2013", fontdict={"fontsize" : 14})
plt.ylabel("Ambulance Diversion Hours\n (Yearly Average)",  fontdict={"fontsize" : 14})

plt.plot([
    pediatric_trauma_df[pediatric_trauma_df['year'] == 2013 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(), 
    pediatric_trauma_df[pediatric_trauma_df['year'] == 2014 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(), 
    pediatric_trauma_df[pediatric_trauma_df['year'] == 2015 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(), 
    pediatric_trauma_df[pediatric_trauma_df['year'] == 2016 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(),
    pediatric_trauma_df[pediatric_trauma_df['year'] == 2017 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(),
    ],
    'b.-'
    )

plt.plot([
    non_pediatric_trauma_df[non_pediatric_trauma_df['year'] == 2013 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(), 
    non_pediatric_trauma_df[non_pediatric_trauma_df['year'] == 2014 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(), 
    non_pediatric_trauma_df[non_pediatric_trauma_df['year'] == 2015 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(),
    non_pediatric_trauma_df[non_pediatric_trauma_df['year'] == 2016 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(),
    non_pediatric_trauma_df[non_pediatric_trauma_df['year'] == 2017 ].EMS_AMB_DIVERS_TOTL_HOURS.mean(),
    ],
    'r.-'
)

plt.tight_layout()

plt.show()
```

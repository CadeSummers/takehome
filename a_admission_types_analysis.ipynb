```python
#2
import pandas as pd
import numpy as np
from time import time
from matplotlib import pyplot as plt
from excel_analysis import a_admission_types

data = a_admission_types()

avail_24hr_dfs = data[0]
avail_oncall_dfs = data[1]

plt.title("Non-urgent EMS Visits & Admissions Compared Between On-Call and 24hr Operation Facilities, YoY \n", fontdict={"fontsize" : 16})

subplot_titles = [
    "Anesthetic Availability\n 24hrs (b) vs On-Call (r)",
    "Lab Services Availability\n 24hrs (b) vs On-Call (r)",
    "Operation Room Availability\n 24hrs (b) vs On-Call (r)",
    "Pharmacy Availability\n 24hrs (b) vs On-Call (r)",
    "Physician Availability\n 24hrs (b) vs On-Call (r)",
    "Psychiatric Emergency Room Availability\n 24hrs (b) vs On-Call (r)",
    "Radiological Services Availability\n 24hrs (b) vs On-Call (r)" 
]


plt.xlabel("Years since 2013", fontdict={"fontsize" : 14})
plt.ylabel("Average Yearly Admissions\n",  fontdict={"fontsize" : 14})


plt.figure(figsize=[10,10])
for i in range(len(avail_24hr_dfs)):
    
    
    plt.subplot(4,2,i+1).set_title( subplot_titles[i] )
    
    
    plt.plot([
        avail_24hr_dfs[i][ avail_24hr_dfs[i]['year'] == 2013 ].EMS_VISITS_NON_URGENT_adm.mean(), 
        avail_24hr_dfs[i][ avail_24hr_dfs[i]['year'] == 2014 ].EMS_VISITS_NON_URGENT_adm.mean(), 
        avail_24hr_dfs[i][ avail_24hr_dfs[i]['year'] == 2015 ].EMS_VISITS_NON_URGENT_adm.mean(), 
        avail_24hr_dfs[i][ avail_24hr_dfs[i]['year'] == 2016 ].EMS_VISITS_NON_URGENT_adm.mean(),
        avail_24hr_dfs[i][ avail_24hr_dfs[i]['year'] == 2017 ].EMS_VISITS_NON_URGENT_adm.mean(),
        ],
        'b.-'
        )



    plt.plot([
        avail_oncall_dfs[i][ avail_oncall_dfs[i]['year'] == 2013 ].EMS_VISITS_NON_URGENT_adm.mean(), 
        avail_oncall_dfs[i][ avail_oncall_dfs[i]['year'] == 2014 ].EMS_VISITS_NON_URGENT_adm.mean(), 
        avail_oncall_dfs[i][ avail_oncall_dfs[i]['year'] == 2015 ].EMS_VISITS_NON_URGENT_adm.mean(),
        avail_oncall_dfs[i][ avail_oncall_dfs[i]['year'] == 2016 ].EMS_VISITS_NON_URGENT_adm.mean(),
        avail_oncall_dfs[i][ avail_oncall_dfs[i]['year'] == 2017 ].EMS_VISITS_NON_URGENT_adm.mean(),
        ],
        'r.-'
    )
    
    plt.tight_layout()

plt.show()

```
